```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
dataset = sb.load_dataset('titanic')
dataset.head()
```

```python
sb.distplot(x=dataset['age'],bins=10)
```

```python
sb.jointplot(x=dataset['age'],y=dataset['fare'],kind='hex')
```

```python
sb.rugplot(dataset['fare'])
```

```python
sb.barplot(x='sex',y='age',data=dataset)
```

```python
sb.countplot(x='sex',data=dataset)
```

```python
sb.boxplot(x='sex',y='age',data=dataset)
```

```python
sb.violinplot(x='sex',y='age',data=dataset)
```

```python
sb.stripplot(x='sex',y='age',data=dataset,jitter=False)
```

```python
sb.swarmplot(x='sex',y='age',data=dataset)
```

```python
corr = dataset.corr()
sb.heatmap(corr)
```
